## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have

```python
def generate_uniform_distribution(bottom, ceiling, count):
    data = np.random.randint(bottom, ceiling, count)
    return data
generate_uniform_distribution(10,60,1000)
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
from scipy.stats import norm 
```

```python
# define our function:

def generate_uniform_norm(mean, std_dev, n):
    data = norm.rvs(loc=mean, scale=std_dev, size=n)
    return data

# case A
mean = 10         
std_dev = 1      
n = 1000          

data_3 = generate_uniform_norm(mean, std_dev, n)


```

```python
# Define parameters
mean = 10         
std_dev = 50      
n = 1000          

# Generate normally distributed numbers using scipy.stats.norm
data_4 = generate_uniform_norm(mean, std_dev, n)
```

```python
# Histogram for random numbers generated with spicy norm
plt.subplot(1, 2, 1)
plt.hist(data_3, bins=100, color='blue', alpha=0.7, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Count')

# Histogram for random numbers generated with spicy norm
plt.subplot(1, 2, 2)
plt.hist(data_4, bins=100, color='orange', alpha=0.7, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Count')

# Show the plots
plt.tight_layout()
plt.show()
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# Define the function to generate exponentially distributed numbers
def generate_exponential_data(mean, n):
    """
    Generate a list of numbers that are exponentially distributed with a given mean.
    
    Args:
        mean (float): The mean (scale) of the exponential distribution.
        n (int): The number of numbers to generate.
        
    Returns:
        list: A list of exponentially distributed numbers.
    """
    # The scale parameter of the exponential distribution is the same as the mean
    scale = mean
    exponential_data = np.random.exponential(scale, n)
    
    return exponential_data

```

```python
# 1. Use the function to generate two number sequences with the size of 10 and 100.

data_5 = generate_exponential_data(10, 10)
data_6 = generate_exponential_data(10,100)

# 2. Plot the data
# Histogram for random numbers generated with spicy norm
plt.subplot(1, 2, 1)
plt.hist(data_5, bins=100, color='blue', alpha=0.7, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Count')
plt.xlim(0,10)

# Histogram for random numbers generated with spicy norm
plt.subplot(1, 2, 2)
plt.hist(data_6, bins=100, color='orange', alpha=0.7, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Count')
plt.xlim(0,100)

# Show the plots
plt.tight_layout()
plt.show()
```

How are the two distributions different?

The mean changes, so the distribution changes as well.
